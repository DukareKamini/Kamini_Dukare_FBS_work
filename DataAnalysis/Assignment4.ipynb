```python
import numpy as np
import pandas as pd
```

### Tasks: 

##### 1. Create a DataFrame: o Load the above dataset into a Pandas DataFrame.

```python
data = pd.DataFrame([
    [1001, "Alice", "Laptop", "Electronics", 2, 800, 1600],
    [1002, "Bob", "Phone", "Electronics", 1, 500, 500],
    [1003, "Charlie", "Chair", "Furniture", 4, 100, 400],
    [1004, "Alice", "Table", "Furniture", 1, 250, 250],
    [1005, "David", "Laptop", "Electronics", 1, 800, 800],
    [1006, "Charlie", "Phone", "Electronics", 2, 500, 1000],
    [1007, "Bob", "Chair", "Furniture", 2, 100, 200],
    [1008, "Alice", "Phone", "Electronics", 1, 500, 500]],
    columns=["Order ID", "Customer", "Product", "Category", "Quantity", "Price", "Total Sales"])
```

```python
data
```

##### 2. Add a New Column: o Add a column Discount that gives a 10% discount on total sales. 

```python
data['Discount'] = data['Total Sales'] * 0.10
```

```python
data
```

##### 3. Add a New Row: o Add a new sales record: Order ID = 1009, Customer = "Eve", Product = "Tablet", Category = "Electronics", Quantity = 1, Price = 300, Total Sales = 300. 

```python
data.loc[8] = [1009, "Eve", "Tablet", "Electronics", 1, 300, 300, 30.0]
```

```python
data
```

##### 4. Filter Data: o Retrieve only the sales records where the Category is "Electronics".

```python
data[data['Category'] == 'Electronics']
```

##### 5. Selection: o Select the Customer and Total Sales columns. 

```python
data[['Customer','Total Sales']]
```

##### 6. Grouping: o Group the data by Customer and calculate the total sales per customer.

```python
data.groupby('Customer')['Total Sales'].sum()
```

##### 7. Sorting: o Sort the DataFrame based on Total Sales in descending order. 

```python
data.sort_values(by='Total Sales', ascending=False)
```

##### 8. Apply Function: o Apply a lambda function to create a new column Sales Category, where values are labeled as "High" if Total Sales > 1000, otherwise "Low". 

```python
data['Sales Category'] = data['Total Sales'].apply(lambda x: 'High' if x > 1000 else 'Low')
```

```python
data['Sales Category']
```

```python
data
```

##### 9. Save DataFrame: o Save the final DataFrame as a CSV file named sales_data.csv.

```python
data.to_csv('sales_data.csv', index=False)
```
