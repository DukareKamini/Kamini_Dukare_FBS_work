```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Refer to the tips.csv dataset. The dataset contains total 244 tips(data about customers served by a restaurant) across 7 different columns.

##### totbill - Total bill (cost of the meal), including tax, in US dollars 
##### tip - Tip (gratuity) in US dollars 
##### sex - Sex of person paying for the meal (0=male, 1=female) 
##### smoker - Smoker in party? (0=No, 1=Yes) 
##### day - Sun, Sat, Thu, Fri 
##### time - Night and Day 
##### size - Size of the party

```python
data = pd.read_csv('Datasets/tips.csv')
```

```python
data.head()
```

##### 1. Plot a day wise boxplot of totbill column. 

```python
plt.figure(figsize=(8,5))
colors = {'Sun': 'blue', 'Sat': 'skyblue', 'Fri': 'green', 'Thur': 'orange'}
sns.boxplot(x='day', y='total_bill', data=data, palette=colors)
plt.title('Day-wise Boxplot of Total Bill')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

##### 2. Plot a pariplot by setting hue parameter t

```python
sns.pairplot(data, hue='sex')
```

##### 3. Draw a scatter plot of tip and totalbill and also allot the points according to the sex

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=data)
```

##### 4. Plot a hitsogram of totbill for 50 bins 

```python
plt.hist(data["total_bill"], edgecolor = 'black', bins=50)
plt.show()
```
